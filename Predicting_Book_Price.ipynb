**Data Preprocessing**

```python
df_data[df_data.isnull().any(axis=1)].count()
```

***The price of a book depends on it's reviews, ratings, genre and book cateogary. Usually when a person buys a book s/he doesn't look at the title, author, edition or synopsis.\
E.g.: If a person likes humorous books s/he tends to buy those types of books.***  

```python
df_data.drop(['Title','Author','Edition','Synopsis'],axis=1,inplace=True)
```

```python
df_data
```

```python
df_data = df_data.sample(frac=1).reset_index(drop=True)
```

```python
df_data
```

```python
df_data.dtypes
```

```python
df_data['Reviews'] = df_data['Reviews'].map(lambda x: x.lstrip('+-').rstrip('out of 5 stars'))
```

```python
df_data['Reviews'] = df_data['Reviews'].astype(float)
```

```python
df_data
```

```python
df_data['Ratings'] = df_data['Ratings'].map(lambda x: x.lstrip('+-').rstrip('customer reviews'))
```

```python
df_data['Ratings']
```

```python
df_data['Ratings'] = df_data['Ratings'].astype(int)
```

```python
df_data.dtypes
```

```python
df_data
```

```python
headers = ['Reviews out of 5','Ratings (customer reviews)','Genre','Book_Category','Price']
df_data.columns = headers
```

```python
df_data
```

```python
print(df_data['Genre'].value_counts())
```

```python
print(df_data['Book_Category'].value_counts())
```

```python
df_data = pd.get_dummies(df_data,columns=['Genre'])
```

```python
df_data
```

```python
df_data = pd.get_dummies(df_data,columns=['Book_Category'])
```
